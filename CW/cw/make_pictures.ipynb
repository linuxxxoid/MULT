```python
import cv2
from os import path
import matplotlib.pyplot as plt
import numpy as np
```

```python
frame = 103 #  

file1 = "frameN/frame_{:d}.jpg".format(frame)
file2 = "frameD/frame_{:d}.jpg".format(frame)

im1 = cv2.cvtColor(cv2.imread(file1),
                   cv2.COLOR_BGR2RGB)
im2 = cv2.cvtColor(cv2.imread(file2),
                   cv2.COLOR_BGR2RGB)
im2 = cv2.resize(im2, (im1.shape[1], im1.shape[0]))

_, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 7), dpi=100)

ax1.set_title("NONE")
ax1.imshow(im1)
ax1.axis(False)

ax2.set_title("DV PAL")
ax2.imshow(im2)
ax2.axis(False)
```

```python
frame = 103

file1 = "frameN/frame_{:d}.jpg".format(frame)
file2 = "frameD/frame_{:d}.jpg".format(frame)

im1 = cv2.cvtColor(cv2.imread(file1),
                   cv2.COLOR_BGR2YUV)
im2 = cv2.cvtColor(cv2.imread(file2),
                   cv2.COLOR_BGR2YUV)
im2 = cv2.resize(im2, (im1.shape[1], im1.shape[0]))

diff = np.int32(im1[:, :, 0]) - np.int32(im2[:, :, 0])
vmax = 15
vmin = -15

plt.figure(figsize=(7, 6), dpi=100)
plt.matshow(diff, interpolation="None", vmin=vmin, vmax=vmax)
plt.colorbar()
plt.title("  : im - imDV", pad=20)
```

```python
frame = 103 #  

file1 = "frameN/frame_{:d}.jpg".format(frame)
file2 = "frameD/frame_{:d}.jpg".format(frame)

im1 = cv2.cvtColor(cv2.imread(file1),
                   cv2.COLOR_BGR2YUV)
im2 = cv2.cvtColor(cv2.imread(file2),
                   cv2.COLOR_BGR2YUV)

num_bins = 20

plt.figure(figsize=(7, 6), dpi=100)
plt.hist(im1[:, :, 0].reshape(-1), num_bins, facecolor='blue', label="NONE")
plt.hist(im2[:, :, 0].reshape(-1), num_bins, facecolor='red', label="DV_PAL", alpha=0.5)#H264
plt.legend()
plt.title(" ")
plt.grid(True)
```

```python
folder = "frameN"
file = "NONE.avi"

vidcap = cv2.VideoCapture(file)

i = 0

while True:
  success, image = vidcap.read()
  
  if success:
      print("write {:d} frame".format(i + 1))
      cv2.imwrite(path.join(folder, "frame_{:d}.jpg".format(i)), image)  # save frame as JPEG file
      i += 1
  else:
      break
```
